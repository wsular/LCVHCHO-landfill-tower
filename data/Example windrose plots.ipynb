# Windrose Plot Examples

## Lewiston-Clarkston Valley Formaldehyde Study (2016)

Requirements:

* [pandas](http://pandas.pydata.org)
* [windrose](https://pypi.python.org/pypi/windrose)

### Basic setup

```python
%matplotlib inline

import pandas as pd
from windrose import WindroseAxes
```

```python
df = pd.read_csv(r'excerpt_23Jun-02Jul_halfhourly.dat',
                 index_col=0, # timestamps first column
                 parse_dates=True, # iso8601
                 header=1, # 1st line metadata 2nd headers
                 skiprows=[2,3], # 3rd units 4th agg type             
                 # Table 129 (p484) in CR1000 manual rev. 4/13/15
                 # https://s.campbellsci.com/documents/us/manuals/cr1000.pdf
                 na_values=['NAN', -7999, 7999], keep_default_na=False)
```

### Scatter Plot

```python
df.WD_10m_unit_Avg.plot(style='k.', alpha=0.33, figsize=(12,4))
```

### Windrose Plots

#### Bar Histogram, Normalized

```python
ax = WindroseAxes.from_ax()
ax.bar(df.WD_10m_unit_Avg, df.WS_10m_Avg, normed=True, bins=range(0,9,1))
ax.set_legend()
```

#### Bar Histogram, Not Normalized

Above plot without normalization.

```python
nx = WindroseAxes.from_ax()
nx.bar(df.WD_10m_unit_Avg, df.WS_10m_Avg, bins=range(0,9,1))
nx.set_legend()
```

#### Box Histogram, Normalized

```python
bx = WindroseAxes.from_ax()
bx.box(df.WD_10m_unit_Avg, df.WS_10m_Avg, normed=True, bins=range(0,9,1))
bx.set_legend()
```

#### Filled Contour, Normalized

```python
cx = WindroseAxes.from_ax()
cx.contourf(df.WD_10m_unit_Avg, df.WS_10m_Avg, normed=True, bins=range(0, 9, 1))
cx.set_legend()
```
